```python
def sexType(value):
    if value=="M":
        return 0
    else:
        return 1
def chestPainType(value):
    if value=='ASY':
        return 0
    elif value=='ATA':
        return 1
    elif value=='NAP':
        return 2
    else:
        return 3
def restingECG(value):
    if value=='LVH':
        return 0
    elif value=='Normal':
        return 1
    else:
        return 2
def exerciseAngina(value):
    if value=='Y':
        return 0
    else:
        return 1
def st_Slope(value):
    if value=='Down':
        return 0
    elif value=='Flat':
        return 1
    else:
        return 2
def dfConversion(newdf):
    newdf.Sex = newdf.Sex.apply(sexType)
    newdf.ChestPainType = newdf.ChestPainType.apply(chestPainType)
    newdf.RestingECG = newdf.RestingECG.apply(restingECG)
    newdf.ExerciseAngina = newdf.ExerciseAngina.apply(exerciseAngina)
    newdf.ST_Slope = newdf.ST_Slope.apply(st_Slope)
    return newdf
```

```python
heart_failure_data=dfConversion(df)
heart_failure_data
```

```python
heart_failure_data.info()
```

```python
#Checking outliers
```

```python
plt.figure(figsize=(18,15))
sns.boxplot(data=heart_failure_data)
```

```python
#from observatation it seems features cholesterol, restingBP and maxHR do have an outlier. Need to remove outlier first
```

```python
#Removing outliers using Z-Score
```

```python
heart_failure_data['Cholessterol_Zscore']=(heart_failure_data.Cholesterol-heart_failure_data.Cholesterol.mean())/heart_failure_data.Cholesterol.std()
heart_failure_data['RestingBP_Zscore']=(heart_failure_data.RestingBP-heart_failure_data.RestingBP.mean())/heart_failure_data.RestingBP.std()
heart_failure_data['MaxHR_Zscore']=(heart_failure_data.MaxHR-heart_failure_data.MaxHR.mean())/heart_failure_data.MaxHR.std()
heart_failure_data.head()
```

```python
df_no_outliers=heart_failure_data[(heart_failure_data.Cholessterol_Zscore < 3) & (heart_failure_data.Cholessterol_Zscore > -3) & (heart_failure_data.RestingBP_Zscore < 3) & (heart_failure_data.RestingBP_Zscore > -3) & (heart_failure_data.MaxHR_Zscore < 3) & (heart_failure_data.MaxHR_Zscore > -3) ]
df_no_outliers
```

```python
df_no_outliers.describe()
```

```python
df_no_outliers.shape
```

```python
plt.figure(figsize=(18,15))
sns.boxplot(data=df_no_outliers)
```

```python
df_no_outliers.drop(['Cholessterol_Zscore','RestingBP_Zscore','MaxHR_Zscore'],axis=1,inplace= True)
df_no_outliers.head()
```

```python
Q1=df_no_outliers.quantile(0.25)
Q3=df_no_outliers.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
#Using IQR for removal of an outlier
```

```python
data =df_no_outliers[~((df_no_outliers < (Q1 - 1.5 * IQR)) |(df_no_outliers > (Q3 + 1.5 * IQR))).any(axis=1)]
data.shape
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler=StandardScaler
```

```python
col_scale=['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']
std_scaler = StandardScaler()
data[col_scale] = std_scaler.fit_transform(data[col_scale])
data=pd.get_dummies(data,columns=['Sex','ChestPainType','FastingBS','RestingECG','ExerciseAngina','ST_Slope'])
data.head()
```

```python
X=data.drop(['HeartDisease'],axis=1)
y=data['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
#finding n-neighbor with more accuracy
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
knn_scores=[]
k_range=range(1,25)
for k in k_range:
        knn_classifier=KNeighborsClassifier(n_neighbors=k)
        scores=cross_val_score(knn_classifier,X_train,y_train,cv=10,scoring='accuracy')
        knn_scores.append(scores.mean())
```

```python
plt.plot(k_range, knn_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
knn_classifier=KNeighborsClassifier(n_neighbors=15)
knn_classifier.fit(X_train,y_train)
score=cross_val_score(knn_classifier,X_train,y_train,cv=10,scoring='accuracy')
score.mean()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
y_predict=knn_classifier.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
#Confusion metrix
cm=confusion_matrix(y_test,y_predict)
plt.title('Confusion Matrix_Heatmap')
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report
print("Classification Report")
print(classification_report(y_test,y_predict))
```
